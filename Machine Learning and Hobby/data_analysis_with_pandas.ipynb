```python
## object to integer

df['Yield (1 000 kg)'] = df['Yield (1 000 kg)'].astype(int)
df['Number of enterprises'] = df['Number of enterprises'].astype(int)
df['Area total (ha)'] = df['Area total (ha)'].astype(int)
df['Young population not yet yielding any crop (ha)'] = df['Young population not yet yielding any crop (ha)'].astype(int)
df['Crop yielding population (ha)'] = df['Crop yielding population (ha)'].astype(int)
df['High tunnel cultivation (ha)'] = df['High tunnel cultivation (ha)'].astype(int)
df['Yield (kg/ha)'] = df['Yield (kg/ha)'].astype(int)
```

```python
EyL=[] 
new_df=[]
Species = []
averages =[]
average_highest_rate =[]
for i in ELY_centre:
    for j in species:
        
        Eyl = df.loc[df['ELY centre']== i]
        SPecies = Eyl.loc[Eyl['species']== j]
        average =SPecies['Yield (1 000 kg)'].mean(axis = 0)
        average_highest_ratee =SPecies['Yield (kg/ha)'].mean(axis = 0)
        EyL.append(Eyl)
        Species.append(SPecies)
        averages.append(average)
        average_highest_rate.append(average_highest_ratee)
        dfnew = {'Center': i,
                'species': j,
                'Average over the years': average,
                'Yield (kg/ha)(mean)': average_highest_ratee}
        new_df.append(dfnew)

                
```

```python
dff = pd.DataFrame(new_df, columns = ['Center', 'species','Average over the years','Yield (kg/ha)(mean)'])
```

```python
dff
                                             
```

```python
## For each ELY centre display the list of the top 3 produce items in terms of Yield (take average over the years)\
```

```python
final_output =[]
for i in ELY_centre:
        Eyl = dff.loc[dff['Center']== i]
        final = Eyl.nlargest(3, ['Average over the years'])
        final_output.append(final)
import numpy as np
final_output =np.array(final_output)
final_output = final_output.reshape(48,4)
final = pd.DataFrame(final_output, columns = ['Center', 'Species','Average over the years','Yield (kg/ha)(mean)'])
final
```

```python
## c) In Satakunta, is there any item(s) that was produced at the highest rate (kg/ha) in any of the years?

```

```python
Max_values =[]
for i in year:
    for j in species:
        Year = df.loc[df['year']== i]
        Year = Year.loc[Year['species']== j]
        max_value = Year.nlargest(1, ['Yield (kg/ha)'])
        Max_values.append(max_value)
```

```python
Max_values =np.array(Max_values)
Max_values = Max_values.reshape(234,10)
max_values = pd.DataFrame(Max_values, columns = ['year', 'ELY centre','species','Number of enterprises',
                                                 'Area total (ha)','Young population not yet yielding any crop (ha)',
                                                 'Crop yielding population (ha)','High tunnel cultivation (ha)',
                                                 'Yield (1 000 kg)','Yield (kg/ha)'])
```
